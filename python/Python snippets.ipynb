````python
## Geocode

```python
import urllib.request
import json

def get_geocode_location(address):
    api_key = ''
    address = address.replace(' ', '+')
    q = f'https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}'
    response = urllib.request.urlopen(q).read()
    response = str(response, 'utf-8') # convert to regular string
    return json.loads(response)

p = get_geocode_location('new benin')

{'results': [
    {'address_components': [{'long_name': 'New Benin II',
     'short_name': 'New Benin II',
     'types': ['administrative_area_level_3', 'political']},
    {'long_name': 'Benin City',
     'short_name': 'Benin City',
     'types': ['locality', 'political']},
    {'long_name': 'Edo',
     'short_name': 'ED',
     'types': ['administrative_area_level_1', 'political']},
    {'long_name': 'Nigeria',
     'short_name': 'NG',
     'types': ['country', 'political']}],
   'formatted_address': 'New Benin II, Benin City, Nigeria',
   'geometry': {'bounds': {'northeast': {'lat': 6.370502, 'lng': 5.6491341},
     'southwest': {'lat': 6.3421161, 'lng': 5.625776500000001}},
    'location': {'lat': 6.353154, 'lng': 5.6379087},
    'location_type': 'APPROXIMATE',
    'viewport': {'northeast': {'lat': 6.370502, 'lng': 5.6491341},
     'southwest': {'lat': 6.3421161, 'lng': 5.625776500000001}}},
   'place_id': 'ChIJF6JSel3TQBAROJJK3QYxW04',
   'types': ['administrative_area_level_3', 'political']}
   ],
 'status': 'OK'}
```

## Annihilation

```python
def annihilate(string):
    """Recursive annihilation of adjacent strings"""

    s = [each.lower() for each in string]
    l = len(s)
    n = ""
    i = 0

    while i < l:
        if i == l-1: # we've gotten to the last index, break loop
            n += string[i]
            break
        else:
            if s[i] == s[i+1]:
                i += 2
            else:
                n += string[i]
                i += 1

    if string != n:
        return annihilate(n)
    print('After total annihilation ', n)
    return n

assert(annihilate('cCAbBb') == 'Ab')
assert(annihilate('AbCcBaCB') == 'CB')
```

## PostgreSQL

```python
import psycopg2 as pg2

# Warning Never, never, NEVER use Python string concatenation (+) or string parameters interpolation (%) to
# pass variables to a SQL query string. Not even at gunpoint.

# Connect to the default postres user
conn = pg2.connect(user='postgres',
                   password='postgres',
                   host='localhost',
                   port='5432',
                   database='postgres',)
conn.autocommit = True # This line is important when creating the database
c = conn.cursor()

cmd = """
CREATE DATABASE my_todo_list
    WITH
    OWNER = postgres
    ENCODING = 'UTF8'
    LC_COLLATE = 'English_United States.1252'
    LC_CTYPE = 'English_United States.1252'
    TABLESPACE = pg_default
    CONNECTION LIMIT = -1;
"""

c.execute(cmd)
conn.commit()

conn = pg2.connect(user='postgres',
                   password='postgres',
                   host='localhost',
                   port='5432',
                   database='my_todo_list',)
c = conn.cursor()

tbl = """
CREATE TABLE list
    (
        id serial PRIMARY KEY,
        created timestamp with time zone NOT NULL,
        item character varying(150)
    );"""

c.execute(tbl)
conn.commit()

add_col = """
ALTER TABLE list
ADD status boolean NOT NULL"""
c.execute(add_col)
conn.commit()

conn.commit()

# Insert items
# We have colums created, item, status

items = [
    ('2019-01-16', "item 1", 'false'),
    ('2019-01-16', "item 2", 'true'),
    ('2019-01-16', "item 3", 'false'),
    ('2019-01-16', "item 4", 'true'),
    ('2019-01-16', "item 5", 'false')
]

q = "INSERT INTO list (created, item, status) VALUES (%s, %s, %s)"

# print(items)

for each in items:
    c.execute(q, each)
conn.commit()
conn.close()
```

## MySQL

```python
import MySQLdb

db=MySQLdb.connect(user='admin',
    passwd="admin",
    host='localhost',
    db="world")
```
````
