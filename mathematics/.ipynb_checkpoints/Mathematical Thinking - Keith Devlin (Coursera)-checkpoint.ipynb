Take the following conditions and answer the following questions

$
n \text{ is a multiple of }5\\
n \text{ is a multiple of }20\\
n \text{ is even and is a multiple of }5\\
n = 100\\
n^2 \text{ is a multiple of }100
$

**Part 1: Which of the following conditions is necessary for the natural number $n$ go be a multiple of 10?**

**Solution**
$\text{Let P be the statement }n \text { is a multiple of } 10$. Each of these conditions can be taken as $Q$. Because we are talking about $Q \text{ being **necessary** for } P$,  this is a question of $P \implies Q$

**Part 2: Which of the following conditions is sufficient for the natural number $n$ go be a multiple of 10?**

**Solution**
$\text{Let P be the statement }n \text { is a multiple of } 10$. Each of these conditions can be taken as $Q$. Because we are talking about $Q \text{ being **sufficient** for } P$,  this is a question of $Q \implies P$


**Part 3: Which of the following conditions is necessary and sufficient for the natural number n to be a multiple of 10?**

**Solution**
$\text{Let P be the statement }n \text { is a multiple of } 10$. Each of these conditions can be taken as $Q$. Because we are talking about $Q \text{ being **necessary** and **sufficient** for } P$,  this is a question of $Q \Longleftrightarrow P$

**Part 1: If the alarm rings, everyone leaves**

**Solution**

$
P: \text{ Alarm rings }Q \text{: everyone leaves}\\
\text{A case of if } P \text{ then }Q \\
\text{Answer: P is the antecedent}
$

**Part 2: Everyone leaves if the alarm rings**

**Solution**

$
P: \text{ Alarm rings }Q \text{: everyone leaves}\\
\text{A case of P if Q }\\
\text{Answer: P is the antecedent}
$

**Part 3: Keith Cycles only if the sun shines**

**Solution**

$
P: \text{ Keith cycles }Q \text{: sun shines}\\
\text{A case of }P \text{ only if }Q\\
\text{Answer: P is the antecedent}
$

**Part 4: Joe leaves whenever Amy arrives**

**Solution**

$
P: \text{ Joe leaves }Q \text{: Amy arrives}\\
\text{A case of }P \text{ whenever }Q\\
\text{Answer: Q is the antecedent}
$
