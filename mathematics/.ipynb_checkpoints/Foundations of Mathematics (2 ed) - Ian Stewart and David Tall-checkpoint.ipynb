## Chapter 2

When any natural number is squared, each of its prime factors appear an even number of times.

Example. $72 = 2 . 2 . 2 . 3 . 3 = 2^3 . 3 ^2\\ 72^2 = (2^3 . 3^2)^2\\ 72^2 = 2^6 . 3^4$

We see that all the powers are even


This can be used to prove that $\surd 2$ is irrational

Assume $\surd 2 = \frac{m}{n}\\ (\frac{m}{n})^2 = \frac{m^2}{n^2}$

Assuming $m, n$ can each be factorized, squaring each will result in an even power for each prime factor. If you were to cancel any factor, $p$, we would have to cancel $p^2$ since each $p$ occur in pair. At the end we will have all even powers for all the remaining factors.

But the prime factorization of $2$ gives an odd power. So it cannot be the case that there is any number $\frac{m}{n}$ whose square is $2$

An infinite decimal is called a *real number*

**Archimedes’ Condition**: Given a positive real number $\epsilon$, there exists a positive integer n such that $\frac{1}{10^n} < \epsilon$, where $n$ is the number of digits after the decimal point.

**Lemma 2.1**: If $\frac{m}{n}$ and $\frac{r}{s}$ are rational, with $\frac{r}{s} \ne 0$, then $\frac{m}{n} + (\frac{r}{s})√2$
is irrational.

**Proposition 2.2**: Between any two distinct rational numbers there exists an irrational number.

**Proposition 2.3**: Between any two distinct irrational numbers there exists a rational number.

**Proof**: Let the irrational numbers be $a, b$ with $a < b$. Consider their decimal expansions, and let the $nth$ decimal place be the first in which they differ. Then

$a = a_0 · a_1 . . . a_{n–1}a_n . . . ,\\ b = a_0 · a_1 . . . a_{n–1}b_n . . . ,$

where $a_n \ne b_n$.

Let $x = a_0 · a_1 . . . a_{n–1}b_n$. Then $x$ is rational and $a < x \le b$.
But since $b$ is irrational, $x \ne b$, so we must have $a < x < b$.

This could be better understood if we use actual digits: Let

$a = 2.58743...., b = 2. 537483...\\ x = 2.58748$

From here we can apply the reasoning above

**Proposition 2.6 (Triangle Inequality)**: If $x$ and $y$ are real numbers, then $|x + y| \le |x| + |y|$.

**Proof** We consider a number of cases

(i)$x \ge 0, y \ge 0$ meaning that $x, y$ are +ve

$|x + y| = x + y = |x| + |y|$

(ii)$x \ge 0, y < 0$ meaning that $x$ is +ve, $y$ is -ve. But here we still need to consider the cases where the size of $x$ is greater than or smaller than that of $y$

(a)value of $x$ larger than that of $y$ eg. $x = 4, y = -2$

Here $x + y \ge 0; 4+(-2) = 2\\ |x + y| = x + y; 4+(-2) = 2\\ x - y = |x| + |y|; 4-(-2) = 6\\ x + y < x - y\\ \therefore |x + y| < |x| + |y|$

(b)value of $x$ smaller than that of $y$ eg. $x = 3, y = -7$

Here $x + y \le 0; 3+(-7) = -4\\ |x + y| = -(x + y); -(3+(-7)) = 4\\ x - y = |x| + |y|; 3-(-7) = 11\\ x + y < x - y\\ \therefore |x + y| < |x| + |y|$

(iii)$x < 0, y \ge 0$ meaning that $x$ is -ve, $y$ is +ve

(iv) $x < 0, y < 0$ meaning that $x, y$ are -ve

Here $|x + y| = -x - y = |x| + |y|$

We can transform $a - \epsilon < x < a + \epsilon$ using the modulus operator

Subract a from all sides $a - \epsilon -a < x - a < a + \epsilon - a\\ -\epsilon < x - a < \epsilon\\ |x-a| < \epsilon$

Since $x - a$ lies between $-\epsilon$ and $+\epsilon$ there is no way that the magnitude of $x - a$ will be greater than $\epsilon$

**Definition 2.7**: A sequence $(a_n)$ of real numbers tends to a limit $l$ if, given any $\epsilon > 0$, there is a natural number $N$ such that $|a_n – l| < \epsilon$ for all $n > N$

**Definition 2.9**: A sequence $(a_n)$ which tends to a limit $l$ is called *convergent*. If no limit exists, it is said to be *divergent*.

**Definition 2.10**: A sequence $(a_n)$ is increasing if each $a_n ≤ a_{n+1}$, so that $a_1 \le a_2 \le a_3 \le...$

**Definition 2.11**: If there exists a real number $k$ such that $a_n < k$ for all $n$ we say that $(a_n)$ is bounded.

**Completeness property of real numbers**:  every bounded sequence of real numbers tends to a real number as limit.
