**Implications: Acronym for antecedents - IWNP (I Will Not Pay)**

Situations where the antecedent comes after the consequent in the statement make use of the following phrases

a. If

b. Whenever

c. Necessary

d. Provided


When the antecedent comes before the consequent in the statement, the following are used

a. Then

b. Implies

c. Only if

d. Sufficient

**TIOS**

$p \Longleftrightarrow q$ is True when $p$ and $q$ have the same Truth value

**Practice 1.1.7** Construct truth tables for the following

\text{(c) }\neg(p \implies q) \Longleftrightarrow [p \wedge (\neg q)]

**Solution**

A statement that is true in all cases is called a **tautology**. When a biconditional (equivalence) is a tautology it means that the two parts are logically equivalent

**Practice 1.1.9** Use the tautologies of Practice 1.1.7 to write out a negation of the following statements

$
\text{(a) Seven is prime or } 2 + 2 = 4\\
\text{(b) If } M \text{ is bounded, then } M \text{ is compact}\\
\text{(c) If roses are red and violets are blue, then I love you.}
$

**Solution**

$
\text{(a) Let }p: \text{ seven is prime; } q: 2 + 2 = 4 \\
\text{Symbolic form } p \vee q\\
\text{This corresponds to case (b) above: "negation of a disjunction"}\\
\text{Negation: } \neg p \wedge \neg q\\

$
\text{(b) Let }p: M \text{ is bounded; } q: M \text{ is compact } \\
\text{Symbolic form } p \implies q\\
\text{This corresponds to case (c) above: "negation of an implication"}\\

\text{Answer: }M \text{ is bounded and } M \text{ is not compact}
$

$
\text{(c) Let }p: \text{ roses are red; } q: \text{ violets are blue } r: \text{ I love you} \\
\text{Symbolic form: } (p \wedge q) \implies r\\
\text{This corresponds to (c) above: "negation of an implication" with a compound antecedent}\\
\text{Negation: } (p \wedge q) \wedge \neg r\\
\text{Answer: Roses are red and violets are blue but I don't love you}
$

**11. Let $p$ be the statement "The figure is a polygon," and let $q$ be the statement "The figure is a circle." Express each of the following statement in symbols.**

a. The figure is a polygon, but it is not a cirle

b. The figure is a polygon or a circle, but not both

c. If the figure is not a circle, then it is a polygon

d. The figure is a circle whenever it is not a polygon

e. The figure is a polygon iff it is not a circle

**Solution**

$
\text{(a)}p \text{ and } \neg q\\
\text{Symbolic: }p \wedge \neg q
$

$
\text{(b) This is a case of **exclusive** OR.}\\
\text{We need to exclude the case where it is **both** from the OR statement}\\
\text{Symbolic: Inclusive OR: } p \vee q\\
\text{Symbolic: When it is both: } p \wedge q\\
\text{Symbolic: excluding the situation where it is both: } (p \vee q) \wedge \neg (p \wedge q)
$

$
\text{(c) If }\neg q \text{, then } p\\
\text{Symbolic: } \neg q \implies p
$

$
\text{(d)}q \text{ whenever } p\\
\text{Symbolic: }p \implies q
$

$
\text{(e)}p \text{ iff } \neg q\\
\text{Symbolic: }p \Longleftrightarrow \neg q
